# Project — Fraud Detection Prediction

## Introduction
Welcome to the fraud detection project for a financial transactions dataset.  
In this project, we aim to develop a predictive model to **identify fraudulent transactions** in financial data.  

This project is crucial for financial institutions to proactively detect and prevent fraud, minimizing financial losses and enhancing security.

---

## Problem Statement
Fraudulent transactions pose a significant challenge for financial institutions, leading to **substantial monetary losses** and **erosion of customer trust**.  

Accurately predicting fraud is essential to mitigate these risks.  

**Objective:**  
Predict whether a transaction is **fraudulent (`isFraud = 1`)** or **legitimate (`isFraud = 0`)**.  

By accurately predicting fraud, institutions can implement **targeted prevention strategies** to reduce fraudulent activities and improve overall security.

---

## Dataset Description
The dataset contains information on **financial transactions**, including both fraudulent and legitimate ones.

**Key specifications:**
- Key dataset specifications: 11142 rows × 10 columns



**Attributes:**

| Attribute        | Description |
|------------------|-------------|
| `step`           | Numerical value representing the time step of the transaction |
| `type`           | Type of transaction (e.g., `TRANSFER`, `CASH_OUT`, `PAYMENT`) |
| `amount`         | Transaction amount |
| `nameOrig`       | Originator's account identifier |
| `oldbalanceOrg`  | Originator's balance **before** the transaction |
| `newbalanceOrig` | Originator's balance **after** the transaction |
| `nameDest`       | Destination's account identifier |
| `oldbalanceDest` | Destination's balance **before** the transaction |
| `newbalanceDest` | Destination's balance **after** the transaction |
| `isFraud`        | Target variable — whether the transaction is fraudulent (`1 = yes`, `0 = no`) |

---

## Type of ML Task
This project is a **supervised classification problem**.  
We aim to build a model that can classify transactions as either:
- **Fraudulent** (`isFraud = 1`)
- **Legitimate** (`isFraud = 0`)

The trained model can then be used by financial institutions to **flag suspicious transactions in real time**.

## 1️⃣ Import Libraries

```python
# Data handling & analysis
import pandas as pd
import numpy as np
# Visualization
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Preprocessing & modeling
from sklearn.preprocessing import StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, precision_score, recall_score, f1_score,
roc_auc_score, classification_report, confusion_matrix)
import joblib
import os
# Display tweaks
sns.set(style='whitegrid')
print('Libraries imported')
```

## 2️⃣ Load Dataset

```python
DATA_PATH = 'Fraud_Analysis_Dataset.csv'
df = pd.read_csv(DATA_PATH)
print('Shape:', df.shape)
df.head()
```

## 3️⃣ Exploratory Data Analysis (EDA)

### 3.1 Basic Checks

```python
print(df.info())
print('\nMissing values per column:\n', df.isnull().sum())
print('\nClass distribution:\n', df['isFraud'].value_counts(normalize=True)*100)
```

### 3.2 Univariate & Bivariate Visualizations

```python
# Transaction type distribution
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='type')
plt.title('Transaction Type Distribution')
plt.show()


# Numerical distributions
num_cols = ['step','amount','oldbalanceOrg','newbalanceOrig','oldbalanceDest','newbalanceDest']
plt.figure(figsize=(12,8))
for i, c in enumerate(num_cols, 1):
    plt.subplot(2,3,i)  # This line and the following lines need to be indented
    sns.histplot(df[c], kde=True)
    plt.title(c)
plt.tight_layout()
plt.show()


# Fraud vs non-fraud by type
plt.figure(figsize=(7,4))
sns.countplot(data=df, x='type', hue='isFraud')
plt.title('Type by Fraud')
plt.show()
```

## 4️⃣ Data Cleaning

### 4.1 Missing Values & Duplicates

```python
print('Missing values:', df.isnull().sum().sum())
print('Duplicates:', df.duplicated().sum())
# df = df.drop_duplicates() # Uncomment if duplicates exist
```

###  4.2 Outlier Detection & Capping

```python
def iqr_outlier_counts(df, col):
    Q1, Q3 = df[col].quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    return ((df[col] < lower) | (df[col] > upper)).sum()
for c in num_cols:
    print(c, 'outliers:', iqr_outlier_counts(df, c))
def cap_outliers_iqr(series):
    Q1, Q3 = series.quantile([0.25,0.75])
    IQR = Q3 - Q1
    lower, upper = Q1 - 1.5*IQR, Q3 + 1.5*IQR
    return series.clip(lower, upper)
for c in num_cols:
    df[c] = cap_outliers_iqr(df[c])
print('Outliers capped (IQR method)')
```

## 5️⃣ Feature Engineering & Preprocessing

### 5.1 Drop Identifiers & Encode Categorical

```python


# Make a copy of the dataframe
df = df.copy()

# Drop 'nameOrig' and 'nameDest' if they exist
cols_to_drop = [col for col in ['nameOrig', 'nameDest'] if col in df.columns]
df.drop(cols_to_drop, axis=1, inplace=True)

# Convert 'type' column to dummy variables (one-hot encoding)
if 'type' in df.columns:
    df = pd.get_dummies(df, columns=['type'], drop_first=True)

# Preview the dataframe
print(df.head())
```

### 5.2 Create New Features

```python
df['balanceOrg_diff'] = df['oldbalanceOrg'] - df['newbalanceOrig']
df['balanceDest_diff'] = df['oldbalanceDest'] - df['newbalanceDest']
print(df.columns)
```

### 5.3 Imputation & Scaling

```python
num_features = [c for c in df.columns if df[c].dtype in ['float64','int64'] and c != 'isFraud']
imputer = SimpleImputer(strategy='median')
scaler = StandardScaler()
df[num_features] = imputer.fit_transform(df[num_features])
df[num_features] = scaler.fit_transform(df[num_features])
print('Imputed and scaled numerical features')
```

## 6️⃣ Train/Test Split

```python
X = df.drop('isFraud', axis=1)
y = df['isFraud']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)
print('Train shape:', X_train.shape, 'Test shape:', X_test.shape)
```

## 7️⃣ Modeling

### 7.1 Logistic Regression (Baseline)

```python
log_clf = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
log_clf.fit(X_train, y_train)
y_pred_log = log_clf.predict(X_test)


print('Logistic Regression')
print(classification_report(y_test, y_pred_log))
print('ROC AUC:', roc_auc_score(y_test, log_clf.predict_proba(X_test)[:,1]))
```

### 7.2 Random Forest (Hyperparameter Search)

```python
rf = RandomForestClassifier(random_state=42, n_jobs=-1)
param_grid = {
'n_estimators': [100, 200],
'max_depth': [None, 10, 20],
'class_weight': ['balanced', 'balanced_subsample']
}
gs = GridSearchCV(rf, param_grid, cv=3, scoring='roc_auc', n_jobs=-1)
gs.fit(X_train, y_train)


print('Best params:', gs.best_params_)
best_rf = gs.best_estimator_
y_pred_rf = best_rf.predict(X_test)


print('Random Forest — Test classification report')
print(classification_report(y_test, y_pred_rf))
print('ROC AUC:', roc_auc_score(y_test, best_rf.predict_proba(X_test)[:,1]))
```

## 8️⃣ Evaluation & Comparison

```python


models = {
'LogisticRegression': (log_clf, y_pred_log, log_clf.predict_proba(X_test)[:,1]),
'RandomForest': (best_rf, y_pred_rf, best_rf.predict_proba(X_test)[:,1])
}


rows = []
for name, (model, y_pred, y_proba) in models.items():
    # Fixed indentation here - all code in the for loop block must be indented
    rows.append({
        'model': name,
        'accuracy': accuracy_score(y_test, y_pred),
        'precision': precision_score(y_test, y_pred),
        'recall': recall_score(y_test, y_pred),
        'f1': f1_score(y_test, y_pred),
        'roc_auc': roc_auc_score(y_test, y_proba)
    })
results_df = pd.DataFrame(rows)
results_df


# Confusion matrix for best model
cm = confusion_matrix(y_test, y_pred_rf)
plt.figure(figsize=(5,4))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues')
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix — Random Forest')
plt.show()
```
